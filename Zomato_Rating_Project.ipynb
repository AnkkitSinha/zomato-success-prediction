```python
def get_Rate(x):
    list_of_review=ast.literal_eval(x)
    if list_of_review is None or len(list_of_review)<= 1:
        return None
    else:
        first_input = list_of_review[0]
        rate = float([i.replace('Rated', '').strip()for i in first_input][0])
        return rate
```
